# Import liberaries

```python
from selenium import webdriver
import pandas as pd
```

# Scrape the website

```python
# specify the path to the selenium webdriver for Chrome browser
driver_path = 'chromedriver_win32\chromedriver.exe'
# create instance from selenium webdriver for Chrome browser
selenium_driver = webdriver.Chrome(driver_path)
# get the data from the website
url = 'https://egypt.souq.com/eg-en/samsung/mobile-phones-33%7Csmart-watches-511%7Ctablets-94%7Cfitness-technology-498%7Cpower-banks-562/samsung/a-t-7/s/?_=1500309575629&sortby=sr&ref=nav&section=2&page=1'
selenium_driver.get(url)
# select the data by specifing the xpath for the required elements
products = selenium_driver.find_elements_by_xpath("//h1[@class='itemTitle']")
prices = selenium_driver.find_elements_by_xpath("//h3[@class='itemPrice']")
```

# Create dataframe from the scraped data

```python
# create lists to store the data
products_lst = [element.text for element in products]
prices_lst = [element.text for element in prices]
```

```python
# create dataframe from the lists
samsung_df = pd.DataFrame({'product':products_lst, 'price':prices_lst})
samsung_df.head()
```

# Save the dataframe to csv file

```python
samsung_df.to_csv('souq_samaung.csv', index=False)
```
